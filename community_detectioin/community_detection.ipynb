```python
import community
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = nx.read_edgelist('epinions-trim.txt',create_using=nx.Graph(),nodetype=int)
print(nx.info(G))
```

```python
def community_detection(G):
    pos = nx.spring_layout(G)
    parts = community.best_partition(G)
    values = [parts.get(node) for node in G.nodes()]
    #print(values)
    size = float(len(set(parts.values())))
    print('no of communities ' + str(size))
    plt.axis("off")
    print('Network with different communities: ')
    nx.draw_networkx(G, pos = pos, cmap = plt.get_cmap("jet"), node_color = values, node_size = 10, with_labels = False)
```

```python
G = nx.read_edgelist('epinions-trim.txt',create_using=nx.Graph(),nodetype=int)
print(nx.info(G))
pos = nx.spring_layout(G)
plt.axis("off")
print('Overall network')
nx.draw_networkx(G, pos = pos, with_labels = False, node_size = 15)
```

```python
community_detection(G)
```
