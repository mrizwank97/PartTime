```python
%matplotlib inline
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from scipy.spatial import distance 
from collections import defaultdict
from scipy.cluster import hierarchy
import math
```

```python
K = nx.read_edgelist('dblp1.txt',create_using=nx.Graph(),nodetype=int)
#K = nx.karate_club_graph()
print(nx.info(K))
```

```python
A = nx.convert_matrix.to_numpy_array(K)
A.shape
```

```python
K.remove_nodes_from(list(nx.isolates(K)))
A = nx.convert_matrix.to_numpy_array(K)

for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        if(i==j):
            temp = 0
            temp1 = 0
            for k in range(j): 
                temp += A[k][j]
            for k in range(j+1,34):
                temp1 += A[k][j]
            A[i][j] = temp+temp1

pij = np.copy(A)
I = np.copy(A)
total = np.sum(A)

for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        pij[i][j] = A[i][j] / total        

pi = np.sum(pij,axis=0)
pj = np.sum(pij,axis=1)

a = 0
b = 0
for i in range(A.shape[0]):
    for j in range(i+1, A.shape[0]):
        for k in range(A.shape[0]):
            pki = pij[k][i]/pi[i] 
            pkj = pij[k][j]/pj[i]
            pkij = (pij[k][i]+pij[k][j])/(pi[i]+pj[j])
            if(pki > 0):
                a = a + pki * (math.log(pki) - math.log(pkij))
            else:
                continue
            if(pkj > 0):
                b = b + pkj * (math.log(pkj) - math.log(pkij))
            else:
                continue
        I[i][j] = pi[i] * a + pj[j]*b
```

```python
similarity = list()
for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        similarity.append(I[i][j])
```

```python
x = [i for i in range(A.shape[0]*A.shape[1])]
plt.scatter(x,similarity)
plt.title('Similarity Grap')
plt.ylabel('Similarity Measure Scatter Plot')
plt.xlabel('Node i * Node j')
plt.show()
```

```python
labels=K.nodes()
Z = hierarchy.average(I)
membership=list(hierarchy.fcluster(Z,t=1.15))
partition = defaultdict(list)
for n,p in zip(list(range(len(K))),membership):
    if(p in partition):
        partition[p].append(labels[n])
partitions = {}
for i in range(len(membership)):
    partitions[i]=membership[i]
```

```python
pos = nx.fruchterman_reingold_layout(K);
plt.figure(figsize=(10,10))
nx.draw_networkx_nodes(K, pos, cmap=plt.cm.RdYlBu, node_color=list(partitions.values()), node_size=50)
nx.draw_networkx_edges(K, pos, alpha=0.5)
plt.axis('off')
plt.show()
```

```python
_SQRT2 = np.sqrt(2) 
def hellinger3(p, q):
    return np.sqrt(np.sum((np.sqrt(p) - np.sqrt(q)) ** 2)) / _SQRT2
```

```python
labels=K.nodes()
path_length=dict(nx.all_pairs_shortest_path_length(K))
n = len(K)
distances=np.zeros((n,n))   
for u,p in path_length.items():
    for v,d in p.items():
        if((u < n and u in K.nodes()) and (v  < n and v in K.nodes())):
            distances[u][v] = d
            distances[v][u] = d
            if u==v: distances[u][u]=0
Y = distance.pdist(distances,hellinger3)
Z = hierarchy.average(Y)
membership=list(hierarchy.fcluster(Z,t=1.15))
partition = defaultdict(list)
for n,p in zip(list(range(len(K))),membership):
    if(p in partition):
        partition[p].append(labels[n])
```
